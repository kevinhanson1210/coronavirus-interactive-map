```python
htmldata.head()
```

```python
for col in htmldata.columns:
    htmldata[col] = htmldata[col].astype(str)
    
htmldata['text'] = htmldata['USAState'] + '<br>' + \
    'Cases ' + htmldata['TotalCases'] + '<br>' + \
    'NewCases ' + htmldata['NewCases'] + '<br>' + \
    'Deaths ' + htmldata['TotalDeaths'] + '<br>' + \
    'New Deaths ' + htmldata['NewDeaths'] + '<br>' + \
    'Active Cases ' + htmldata['ActiveCases'] 

fig2 = go.Figure(data = go.Choropleth(
    locations=htmldata['code'],
    z=htmldata['TotalCases'].astype(float),
    locationmode='USA-states',
    colorscale='Reds',
    autocolorscale=False,
    text=htmldata['text'],
    colorbar_title= '#'
))

fig2.update_layout(
    title_text='2020 Coronavirus Cases by State<br>(Hover for breakdown)',
    geo = dict(
        scope='usa',
        projection=go.layout.geo.Projection(type = 'albers usa'),
        showlakes=True, # lakes
        lakecolor='rgb(255, 255, 255)'),
)
```
